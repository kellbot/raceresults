```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import glob

tenk_headers = ["Group Place","Name","Bib","Age","Place","Gender Place",
                "5K Split","Clock Time","Net Time","Hometown"]
philly_headers = ["Place","Bib","First Name","Last Name","Gender","Age",
                  "Clock Time","Net Time","Pace","Sex total","Sex Place",
                  "Div Total","Div Place","Division"]
half_runners = ["Group Place","Name","Bib","Age","Place","Gender Place",
                "5K Split","10K Split","15K Split","Clock Time","Net Time","Hometown"]
full_runners = ["Group Place","Name","Bib","Age","Place","Gender Place",
                "5M Split","10M Split","Half Split","20M Split","Clock Time","Net Time","Hometown"]


paths = [{'distance': 6.2, 'paths': glob.glob('10k/*.csv'), 'headers': tenk_headers, 'Name': '10K'},
         {'distance': 6.2, 'paths': glob.glob('Philly10k/*.csv'), 'headers': philly_headers, 'Name': 'Philly10K'},
         {'distance': 13.1, 'paths': glob.glob('Half/*.csv'), 'headers': half_runners, 'Name':'Half'},
         {'distance': 26.2, 'paths': glob.glob('Full/*.csv'), 'headers': full_runners, 'Name':'Full'}]


run_data = []

for row in paths:
    for path in sorted(row['paths']):
        tdata = pd.read_csv(path, names=row['headers'])
        tdata['Name'] = row['Name']
        tdata['distance'] = row['distance']
        run_data.append(tdata.copy())

run_df = pd.concat(run_data, axis=0, ignore_index=True)
```

Converting the data to the `np.timedelta64` will make things MUCH easier later on.

```python
def new_timedelta(val):
    
    try:
        parts = val.split(':')
    except AttributeError:
        #np.nan will be here
        return np.nan
    tot = np.timedelta64(0, 'h')
    order = ['seconds', 'minutes', 'hours']
    for o, p in zip(order, parts[::-1]):
        tot += pd.to_timedelta(p + ' %s' % o)
    return tot
    
delta_cols = ['10K Split', '10M Split', '15K Split', '20M Split',
               '5K Split', '5M Split', 'Net Time', 'Clock Time']

for col in delta_cols:
    run_df[col] = run_df[col].map(new_timedelta)
    
```

```python
run_df['StartOffsetMin'] = ((run_df['Clock Time'] - run_df['Net Time'])/np.timedelta64(1, 'm'))

#assume all races started at '12:00 on 1/1/1'
run_df['StartTime'] = pd.to_datetime('12:00 am 1/1/2001') + (run_df['Clock Time'] - run_df['Net Time'])
```

```python
def min_aggregate(df):
    
    df.index = df.index.droplevel()
    return df.resample('45s', how='sum')


uni_times = run_df.groupby(['Name', 'StartTime'])['Clock Time'].count()
min_counts = uni_times.groupby(level='Name').apply(min_aggregate)
min_counts = min_counts.fillna(0)
min_counts
```

```python

with plt.xkcd(length=100):
    fig, axs = plt.subplots(4, 1, figsize=(10,10))

    for ax, (key, ser) in zip(axs.flatten(), min_counts.groupby(level='Name')):
    
        ser.ix[key].plot(ax=ax, grid=False)
        ax.set_title(key)
        ax.set_ylabel('# Runners')
    
    fig.tight_layout()
    
    
```

Just a simple algorithm to find the Heat-number from the count-Series. This method looks for gaps of zeros and then increments the heat-number. This assumes that they're aren't any short 0 gaps within each heat ... however, based on the graph above it doesn't look like that happens.

```python
heat_num = 0
test_min_counts = [0,0,0,1,1,5,2,5,0,0,0,0,0,8,9,7,9,8,7,8,0,0,0,0,0,6,8,9,7]

for cur_num, next_num in zip(test_min_counts, test_min_counts[1:]+[0]):
    
    if (cur_num == 0) and (next_num > 0):
        heat_num += 1
        print 'next heat'
    if cur_num != 0:
        print cur_num, heat_num

    
```

```python
def guess_heat(count_ser):
    
    count_ser.index = count_ser.index.droplevel()
    res = []
    heat_num = 1
    for cur_num, next_num in zip(list(count_ser), list(count_ser)[1:]+[0]):
        res.append(heat_num if cur_num != 0 else np.nan)
        if (cur_num > 0) and (next_num == 0):
            heat_num += 1
        
    return pd.Series(res, index=count_ser.index)


heat_nums = min_counts.groupby(level='Name').apply(guess_heat)
```

```python
heat_nums.ix['10K']
```

```python
with plt.xkcd(length=100):
    fig, axs = plt.subplots(4, 1, figsize=(10,10))

    for ax, (key, ser) in zip(axs.flatten(), heat_nums.groupby(level='Name')):
    
        ser.ix[key].plot(ax=ax, grid=False)
        
        ax.set_title(key)
        ax.set_ylabel('Heat #')
    
    fig.tight_layout()
```

Now to get the heat numbers back into the oringal DataFrame. This is where pandas is AWESOME because has an `.asof` operator which can merge things that have become a wibbly-wobly bowl of goup.
